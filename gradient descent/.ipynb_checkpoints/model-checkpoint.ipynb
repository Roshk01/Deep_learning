# Implement Gradient descent from scratch 

```python
def sigmoid_np(x):
    return 1 / (1 + np.exp(-x))
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def gradient_descent(age,affor,y_true, epochs, loss_thresold):
    w1 = w2 = 1
    bias = 0
    learning_rate = 0.01
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * w2 + affor + bias
        y_predict = sigmoid_np(weighted_sum)
        loss = log_loss(y_true, y_predict)
        
        w1d = (1/n)*np.dot(np.transpose(age),(y_predict-y_true)) 
        w2d = (1/n)*np.dot(np.transpose(affor),(y_predict-y_true)) 

        bias_d = np.mean(y_predict-y_true)
        w1 = w1 - learning_rate * w1d
        w2 = w2 - learning_rate * w2d
        bias = bias - learning_rate * bias_d
        
        print(f'epochs: {i}, w1: {w1}, w2: {w2}, bias: {bias}, loss: {loss}')
        
        if loss<=loss_thresold:
            break
    return w1, w2, bias
    
```

```python
gradient_descent(x_train_scaled['age'],x_train_scaled['affordibility'],y_train,1000, 0.4631)
```
